```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv("student-performance-data.csv")
```

```python
# See how many unique types of schools are there
unique_schools = df['school'].nunique()
print("Number of unique school types:", unique_schools)
```

```python
# If you also want to see the actual school names
school_types = df['school'].unique()
print("Types of schools:", school_types)
```

```python
# If you want counts of each school type
school_counts = df['school'].value_counts()
print("Counts of each school type:\n", school_counts)
# Print the total number of rows (excluding header)
total_rows = len(df)
print("\nTotal number of rows:", total_rows)

# Or, if you want total rows counted from the 'school' column only (ignoring NaN)
total_schools = df['school'].count()
print("Total rows with school values:", total_schools)
```

```python
# 1. First 5 rows (quick look)
print("Head of data:\n", df.head(), "\n")

# 2. Shape (rows, columns)
print("Shape of dataset (rows, columns):", df.shape, "\n")

# 3. Column names
print("Column names:", df.columns.tolist(), "\n")

# 4. Data types of each column
print("Data types:\n", df.dtypes, "\n")

# 5. Summary statistics (only numeric columns by default)
print("Summary statistics:\n", df.describe(), "\n")

# 6. Summary statistics including categorical columns
print("Full summary statistics:\n", df.describe(include='all'), "\n")

# 7. Missing values
print("Missing values in each column:\n", df.isnull().sum(), "\n")

# 8. Unique values count for each column
print("Unique values in each column:\n", df.nunique(), "\n")

# 9. Memory usage
print("Memory usage:\n")
df.info()

# 10. Correlation between numeric columns
print("\nCorrelation matrix:\n", df.corr(numeric_only=True), "\n")
```
